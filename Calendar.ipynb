```python

def is_leap_year(year) :
  """
  Check whether the given year is a leap year or a common year.
  Arguments :
    year -- year, int
  Returns :
    True -- If the year is leap year
    False -- If the year is common year
  """
  if(year % 4 != 0) :
    return False
  elif((year % 4 == 0) & (year % 100 != 0)) :
    return True
  elif((year % 100 == 0) & (year % 400 != 0)) :
    return False
  elif((year % 400 == 0) & (year % 3200 != 0)) :
    return True
  else :
    return False

# Test Program
for iter in range(0, 2):
  year = input("Enter Year:")
  yearType = 'Common Year'
  if(is_leap_year(int(year))) :
      yearType = 'Leap Year'
  print(f'{year} is a {yearType}')

```

```python
def day_of_the_week_julian_int(q, m, k, j) :
  """
  Calculate Julian week day and return integer value(0:Sat, 6:Fri).
  Arguments : 
  q -- day of the month, int
  m -- month, int
  k -- year of the century, int
  j -- zero-based century, int

  Returns :
  julian day of the week -- 0 : Saturday & 6 : Friday
  """
  h = ( q + ((13*(m + 1)) // 5) + k + (k // 4) + (j // 4) - 2*j ) % 7
  return h

def day_of_the_week_iso_int(year, month, day) : 
  """
  Calculate ISO week day and return integer value(1:Mon, 7:Sun).
  Arguments : 
  year -- year, int
  month -- month, int
  day -- day of the month, int

  Returns :
  ISO day of the week -- 1 : Monday & 7 : Sunday
  """
  q = day
  m = month
  y = year
  if(m == 1 or m == 2) :
    m += 12
    y -= 1
  
  k = y % 100 
  j = y // 100

  h = day_of_the_week_julian_int(q, m, k, j)
  d = ((h + 5) % 7) + 1
  return d

def day_of_the_week_iso_str(year, month, day) :
  """Calculate ISO week day and return string value(Mon, Tue, ..)."""
  d = day_of_the_week_iso_int(year, month, day)
  days = { 1 : 'Monday',
           2 : 'Tuesday',
           3 : 'Wednesday',
           4 : 'Thursday',
           5 : 'Friday' ,
           6 : 'Saturday',
           7 : 'Sunday'
        }

  day = days[d]
  return day

# Test Program
print(f'29th January 2022 is {day_of_the_week_iso_str(2022, 1, 29)}')
print(f'8th March 2022 is {day_of_the_week_iso_str(2022, 3, 8)}')
```
