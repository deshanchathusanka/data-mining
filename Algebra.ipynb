```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
class SimultaneousEquation :
  """Matrix operations for semultaneous equations."""
  a = np.zeros([2, 2], dtype = int)
  b = np.array(3, dtype = int)

  def __init__(self, a, b) :
    self.a = a
    self.b = b
  
  def direct_solve(self) :
    """Solve semultaneous equation using in-built function and return solution."""
    solution = np.linalg.solve(a, b) # solve simultaneous equations
    np.allclose(np.dot(a, solution), b) # verify solution
    return solution

  def inverse_solve(self) :
    """Solve semultaneous equation using inverse function and return solution."""
    ainv = np.linalg.inv(a)
    np.allclose(np.dot(a, ainv), np.eye(2)) # verify inverse
    solution = np.dot(ainv, b)
    return solution
  
  def visualize(self) :
    """Visualize soluation in vector form"""
    # create canvas
    plt.figure() 

    # lables
    plt.xlabel('X')
    plt.ylabel('Y')

    # limits
    plt.xlim([-2, 8])
    plt.ylim([-2, 8])

    # feature vectors
    v1 = a[:, 0]
    v2 = a[:, 1]

    # solutions
    solution = self.direct_solve()
    x = solution[0]
    y = solution[1]

    # scaled feature vectors
    v1_scaled = v1*x
    v2_scaled = v2*y

    # plot feature vectors
    plt.quiver(0, 0, v1[0], v1[1], angles = 'xy', scale_units = 'xy', color = 'blue', scale = 1)
    plt.quiver(0, 0, v2[0], v2[1], angles = 'xy', scale_units = 'xy', color = 'orange',scale = 1)

    # plot scaled feature vectors
    plt.quiver(0, 0, v1_scaled[0], v1_scaled[1], angles = 'xy', scale_units = 'xy', color = 'green', scale = 1)
    plt.quiver(v1_scaled[0], v1_scaled[1], v2_scaled[0], v2_scaled[1], angles = 'xy', scale_units = 'xy', color = 'red', scale = 1)

    # plot linear combination of feature vectors
    lin_combination = v1_scaled + v2_scaled
    plt.quiver(0, 0, lin_combination[0], lin_combination[1], angles = 'xy', scale_units = 'xy',color = 'purple' , scale = 1)


    plt.title('Linear Algebra Representation')
    plt.grid() # enable grid
    plt.legend(['VA1, The First Column Of Vector A',
                'VA2, The Second Column Of Vector A',
                'VA1 Scaled By X[0]',
                'VA2 Scaled By X[1]',
                'Linear Combination Of VA1 and VA2'], 
              fontsize = 11) 

    plt.show()

# Test Program
a = np.array([[2, -1], [-1, 2]])
b = np.array([0, 3])
semultaneous_equations = SimultaneousEquation(a, b)

solution = semultaneous_equations.direct_solve()
print(f'Solver Method : \n{solution}\n')
solution = semultaneous_equations.inverse_solve()
print(f'Inverse Method: \n{solution}\n')
semultaneous_equations.visualize()



```
