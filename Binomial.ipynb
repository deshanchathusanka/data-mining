```python
import random as rdm
import numpy as np
import matplotlib.pyplot as plt
```

```python
class BallDrawingGame :
  """
  Implementation of the ball drawing game.
  """
  all_balls = ['red','red','red','black','black','black','black','black','black','black']
  n_draws = 10
  n_trials = 10000
  

  def __init__(self, all_balls, n_draws, n_trials) :
    """Constructor"""
    self.all_balls = all_balls
    self.n_draws = n_draws
    self.n_trials = n_trials

  def __draw_ball(self) :
    """
    Draw single ball and return True if the selected one is black.

    Arguments -- No

    Returns 
    True -- If the ball drawn is black
    False -- Else
    """
    rdm.shuffle(self.all_balls)
    selected_ball = rdm.choice(self.all_balls)

    if(selected_ball == 'black') :
      return True
    else : 
      return False


  def __play_single_trial(self) :
    """
    Play single trial of the game and return number of success for the relevant trial.

    Arguments : No

    Returns :
    number_of_success -- Number of success counts, int
    """
    number_of_success = 0

    for x in range(0, self.n_draws) :
      is_success = self.__draw_ball()
      if(is_success) :
        number_of_success += 1
    
    return number_of_success


  def play(self) :
    """
    Play the game multiple times and return number of success in each trial as a list.
    """
    success_cnt = np.zeros((self.n_trials, 1), dtype = np.int32)

    for x in range(0, self.n_trials) :
      number_of_success = self.__play_single_trial()
      success_cnt[x] = number_of_success
    return success_cnt


# Test Program
n_draws = input("Enter the number of draws : ")
n_trials = input("Enter the number of trials : ")
all_balls = ['red','red','red','black','black','black','black','black','black','black']
ball_drawing_game = BallDrawingGame(all_balls, int(n_draws), int(n_trials))
success_cnt = ball_drawing_game.play()

plt.figure()
bins = int(int(n_draws)/2)
plt.xlabel('Success count')
plt.ylabel('Frequency')
plt.hist(success_cnt, bins)
plt.show()
```
